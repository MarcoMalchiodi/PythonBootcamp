```python
# Standard Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scatter Plot
# a bunch of data scattered all over the chart

X_data = np.random.random(50)*100
Y_data = np.random.random(50)*100
plt.scatter(X_data,Y_data)
plt.show()
```

```python
# If there is a lot of data overlapping, we can use the alpha argument:

X_data = np.random.random(1000)*100
Y_data = np.random.random(1000)*100
plt.scatter(X_data,Y_data, marker="*",alpha=0.3)
plt.show()
```

```python
# Line Plots
years = [1600 + x for x in range(20)]
import random
weights = [random.randint(0,100) for x in range(20)]

plt.plot(years,weights,linestyle="--") # the line chart is the standard one
plt.xticks(years, [int(year) for year in years], rotation=45)  # Set x-axis ticks as integers
plt.show()
```

```python
# Bar Plots
candidates = ["Balubba","Scrofa Assassina", "A' Romano","Perkele"]
votes = [12,34,99,25]

plt.bar(candidates,votes)
plt.show()
```

```python
# Histograms
ages = np.random.normal(20,7.5,1000) # 7.5 is the standard deviation
plt.hist(ages, cumulative=False)
```

```python
# Pie Charts
candidates = ["Balubba","Scrofa Assassina", "A' Romano","Perkele"]
votes = [12,34,99,25]
explodes = [0.3,0,0,0]

plt.pie(votes,labels=candidates,explode=explodes,autopct="%.2f%%",pctdistance=0.7)
```

```python
# Boxplots
heights = np.random.normal(172,8,300)
plt.boxplot(heights)
```
