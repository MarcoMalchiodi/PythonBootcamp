```python
#prime number chekcer
def prime_checker(num):
    counter = 0
    for x in range(2,num):
        if (num%x==0) & (num!=x):
            counter += 1
    if counter > 0:
        print("It's not a prime number")
    else:
        print("It's a prime number")

n = int(input("Insert your number"))

prime_checker(n)
```

```python
#Caeser Code
def kaiser(text, shift, direction):
    list_text = list(text)
    rev_alph = alphabet[::-1]

    if direction == 'encode':
        for i, x in enumerate(list_text):
            if x not in alphabet:
                x = x
            else:
                x = alphabet[((alphabet.index(x))+shift)%len(alphabet)]
                list_text[i] = x
    else:
        for i, x in enumerate(list_text):
            if x not in rev_alph:
                x = x
            else:
                x = rev_alph[((rev_alph.index(x))+shift)%len(rev_alph)]
                list_text[i] = x

    print(''.join(list_text))
    
    

alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
question = 'yes'

while question == 'yes':
    directiona = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    texto = input("Type your message:\n").lower()
    shifto = int(input("Type the shift number:\n"))
    kaiser(texto,shifto,directiona)
    question = input("Do you wish to continue? Type 'yes' or 'no'")
```

```python
#Calculator

def add (n1,n2):
    return n1+n2
def subtract(n1,n2):
    return n1-n2
def multiply(n1,n2):
    return n1*n2
def divide(n1,n2):
    return n1/n2

calc = {
    "+":add,
    "-":subtract,
    "*":multiply,
    "/":divide,
}

def calculator():
    num1 = float(input("What's the first numer?"))
    for key in calc:
        print(key)
    operation = input("Pick an operation")
    num2 = float(input("What's the second number?"))

    function = calc[operation]
    answer = function(num1,num2)

    print(f"{num1} {operation} {num2} = {answer}")

    cont = input("Type 'y' to continue, 'q' to quit, 'n' for a new operation")

    while cont == 'y':
        operation = input("Pick a new operation: ")
        num3 = float(input("Type your next number"))
        function = calc[operation]
        new_answer = function(answer,num3)
        print(f"{answer} {operation} {num3} = {new_answer}")
        answer = new_answer
        cont = input("Do you wish to continue? y/q/n")
    if cont == 'n':
        calculator()


calculator()
```

```python
############### Our Blackjack House Rules #####################

## The deck is unlimited in size. 
## There are no jokers. 
## The Jack/Queen/King all count as 10.
## The the Ace can count as 11 or 1.
## Use the following list as the deck of cards:
## cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
## The cards in the list have equal probability of being drawn.
## Cards are not removed from the deck as they are drawn.
## The computer is the dealer.
import random

def end_game(player,computer):
    player_counter = 0
    computer_counter = 0
    for x in player:
        player_counter += x
    for n in computer:
        computer_counter += n
    
    if player_counter > 21:
        print(f"The computer's hand is {computer} with a score of {sum(computer)}")
        return "You lose!"
    p_score = abs(player_counter -21)
    c_score = abs(computer_counter -21)
    print(f"The computer's hand is {computer} with a score of {sum(computer)}")

    if c_score > p_score or c_score > 21:
        print("You win!")
    elif c_score == p_score:
        print("It's a tie!")
    else:
        print("You lose!")


cards = [11,2,3,4,5,6,7,8,9,10,10,10,10]
fin_question = 'y'

while fin_question == 'y':
    my_hand = [cards[random.randint(0,12)],cards[random.randint(0,12)]]
    pc_hand = [cards[random.randint(0,12)],cards[random.randint(0,12)]]

    print(f"Your hand is {my_hand} with a score of {sum(my_hand)}")
    hidden_c = pc_hand.copy()
    hidden_c[0]='?'
    print(f"The computer's hand is {hidden_c}")
    p_choice = input("Do you wish to draw another card? y/n")

    while p_choice == 'y':
        my_hand.append(cards[random.randint(0,12)])
        print(f"Your new hand is {my_hand} with a score of {sum(my_hand)}")
        p_choice = input("Do you wish to draw another card? y/n")

    for a in range(len(my_hand)):
        if my_hand[a] == 11:
            ace=input("Yoy have an ace, which value do you choose: 1 - 11")
            my_hand[a] = int(ace)
            print(f"Your new hand is {my_hand} with a score of {sum(my_hand)}")

    p_counter = 0
    for c in pc_hand:
        p_counter += c
    if p_counter < 17:
        pc_hand.append(cards[random.randint(0,12)])


    end_game(my_hand,pc_hand)
    fin_question = input("Do you wish to start a new game? y/n\n\n\n")
```

```python
#Coffee Machine
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}


def report():
    for key, value in resources.items():
        print(f"{key}:{value}")

def espresso():
    global total_geld
    global profit
    espro = MENU["espresso"]
    espro_p = espro["cost"]
    espro = espro["ingredients"]
    if total_geld >= espro_p:
        if resources["water"] >= espro["water"] and resources["coffee"] >= espro["coffee"]:
            resources["water"] = resources["water"] - espro["water"]
            resources["coffee"] = resources["coffee"] - espro["coffee"]
            total_geld = total_geld - espro_p
            print(f"Your change is {round(total_geld)}$")
            print("Enjoy your espresso!")
            profit += espro_p
            
        else:
            print("There aren't enough ingredients!")
    else:
        print("You don't have enough money!")

def latte():
    global total_geld
    global profit
    latto = MENU["latte"]
    latto_p = latto["cost"]
    latto = latto["ingredients"]
    if total_geld >= latto_p:
        if resources["water"] >= latto["water"] and resources["coffee"] >= latto["coffee"] and resources["milk"] >= latto["milk"]:
            resources["water"] = resources["water"] - latto["water"]
            resources["milk"] = resources["milk"] - latto["milk"]
            resources["coffee"] = resources["coffee"] - latto["coffee"]
            total_geld = total_geld - latto_p
            print(f"Your change is {round(total_geld)}$")
            print("Enjoy your latte!")
            profit += latto_p
        else:
            print("There aren't enough ingredients!")
    else:
        print("You don't have enough money!")

def cappuccino():
    global total_geld
    global profit
    cappu = MENU["cappuccino"]
    cappu_p = cappu["cost"]
    cappu = cappu["ingredients"]
    if total_geld >= cappu_p:
        if resources["water"] >= cappu["water"] and resources["coffee"] >= cappu["coffee"] and resources["milk"] >= cappu["milk"]:
            resources["water"] = resources["water"] - cappu["water"]
            resources["milk"] = resources["milk"] - cappu["milk"]
            resources["coffee"] = resources["coffee"] - cappu["coffee"]
            total_geld = total_geld - cappu_p
            print(f"Your change is {round(total_geld)}$")
            print("Enjoy your cappuccino!")
            profit += cappu_p
        else:
            print("There aren't enough ingredients!")
    else:
        print("You don't have enough money!")



profit = 0
operating = True


while operating:
    choice = input(print("Please, select an option: espresso/latte/cappuccino"))
    if choice == 'report':
        report()
        print(f"Our total profits are {round(profit)}$")
    elif choice == 'latte' or choice == 'cappuccino' or choice == 'espresso':
        quarters = (int(input("How many quarters?")))*0.25 
        dimes = (int(input("How many dimes?")))*0.10
        nickels = (int(input("How many nickels?")))*0.05
        pennies = (int(input("How many pennies?")))*0.01
        total_geld = quarters+dimes+nickels+pennies

        if choice == 'espresso':
            espresso()
        elif choice == 'latte':
            latte()
        elif choice == 'cappuccino':
            cappuccino()
    else:
        operating = False
```
