```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('002 QueryResults.csv', names=['DATE', 'TAG', 'POSTS'], header=0)
data.DATE = pd.to_datetime(data.DATE)
pivoted_data = data.pivot(index='DATE',columns='TAG', values='POSTS')
pivoted_data.fillna(0,inplace=True)
```

```python
plt.plot(pivoted_data.index, pivoted_data.java)
```

```python
plt.figure(figsize=(16,10))
plt.plot(pivoted_data.index, pivoted_data.java)
```

```python
# Now we can add labels. Also, we're never going to get less than 0 posts, so let's set a lower limit of 0 for the y-axis with .ylim().

plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
```

```python
plt.plot(pivoted_data.index, pivoted_data.java, pivoted_data.python)
```

```python
# But what if we wanted to plot all the programming languages on the same chart? We don't want to type out .plot() a million times, right? We can actually just use a for-loop:

for column in pivoted_data.columns:
    plt.plot(pivoted_data.index, pivoted_data[column])
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)

for column in pivoted_data.columns:
    plt.plot(pivoted_data.index, pivoted_data[column],
             linewidth=3, label=pivoted_data[column].name)
```
