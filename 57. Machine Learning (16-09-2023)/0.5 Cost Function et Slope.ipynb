```python
import numpy as np
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def NN(m1,m2,w1,w2,b):
  z = m1 * w2 + m2 * w2 + b # bias
  return sigmoid(z)
```

```python
# making random weights
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()
```

# **COST FUNCTION**






Takes our data + predictions and returns a number (ex. 0.75).

This number tells us how far our predictions is from the actual data.

Our aim is to minimize this number as much as possible.

(prediction - actual target)**2 > Squared Error Cost Function

The only way to change the prediction is by tweaking the weights and biases.

Screenshot 2023-10-19 225428.png



Let ( b - target )**2, how do we know when to increase/decrease b?

Cosnidering that the function is a parabola, when the slope faces downwards we want to increase, otherwise the opposite.

**Slope of the Cost Function:**
slope(b) must go towards 0. How do I we achieve it?

I) Numerically

II) Algebraically

III) With calculus

We know that slope = rise/run.
Therefore slope = (cost(b+h) - cost(b)) / h
The smaller h's are better at approximation.

If cost(b) = (b-4)**2
then slope = 2(b-4)

```python
def cost(b):
  return (b-4)**2

cost(4) # 4 is the target
```

```python
def numerical_slope(b):
  h = 0.0001
  return ((cost(b+h) - cost(b)) / h)
```

```python
numerical_slope(3) # below the target is negative
numerical_slope(5) # above is positive
```

```python
def slope(b):
  return 2 * (b-4)
```

```python
print(slope(3))
print(slope(5))
```

```python
b = 8
```

```python
b = b - .1 * slope(b)
b # the more we calculate it, the closer we get to our target
```
