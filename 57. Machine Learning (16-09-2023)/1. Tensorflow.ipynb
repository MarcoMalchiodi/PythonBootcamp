```python
string = tf.Variable('This is a string',tf.string)
number = tf.Variable(324,tf.int16) # the most common type
floating = tf.Variable(3.14,tf.float64)
# scaler value > one value only
# vector > more values
```

```python
# Rank/Dregree > number of dimensions involved in a tensor
rank1_tensor = tf.Variable(["Test","ok","bye"],tf.string) # arrays are always at least rank 1
rank2_tensor = tf.Variable([['Test','ok','bye'],['Test','ok','bye']],tf.string)
```

```python
tf.rank(rank2_tensor)
```

```python
# Shape
# numbe rof items in each dimension
rank2_tensor.shape
```

```python
rank1_tensor.shape
```

```python
# Changing Shape
# ex. throwing all 6 elements in rank2 into a rank1
reshaped_tensor = tf.reshape(rank2_tensor,[6])
print(reshaped_tensor)
# or
reshaped_tensor = tf.reshape(rank1_tensor,[3,-1]) # same as [3,1]
print(reshaped_tensor)
```
