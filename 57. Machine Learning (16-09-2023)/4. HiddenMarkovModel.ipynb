```python
import tensorflow_probability as tfp
import tensorflow as tf
```

```python
# WEATHER MODEL
# 1. Cold days are encoded by a 0 and warm days by a 1
# 2. The first day in our sequence has an 80% chance of being cold
# 3. A cold day has a 30% chance of being followed by a warm day
# 4. A warm day has a 20% chance of being followed by a cold day
# On each day the temperature is normally distributed with mean and standard deviation of 0 and 5 (-5 to 5) on a cold day
# and a mean and standard deviation of 15 and 10 (5 to 25)
```

```python
tfd = tfp.distributions # making a shortcut for the following code
initital_distribution = tfd.Categorical(probs=[0.8,0.2]) # refer to point 2 above
transition_distribution = tfd.Categorical(probs=[[0.7,0.3],[0.2,0.8]]) # refer to point 3 and 4

observation_distribution = tfd.Normal(loc=[0. ,15. ], scale=[5. ,10. ]) # refer to point 5

# the loc argument rapresents the mean (0 on a cold day, 15 on a hot day) and the scale is the std deviation
# Categorical is a way to do the distribution
```

```python
model = tfd.HiddenMarkovModel(
    initial_distribution = initital_distribution,
    transition_distribution = transition_distribution,
    observation_distribution = observation_distribution,
    num_steps = 7
)
# steps is the number of steps we want to predict the average remperature for
```

```python
mean = model.mean()

# create a new session in tensorflow to print it out
with tf.compat.v1.Session() as sess:
  print(mean.numpy())
```
