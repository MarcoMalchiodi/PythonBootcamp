```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 1. Prepare the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,4)) # width, height

# 3. Plot data
ax.plot(x,y)

# 4. Cutomize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show
# fig.savefig('path name')
plt.show()
```

Line Plot

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(5,4))

plt.plot(x, x**2)
```

Scatter Plot

```python
fig, ax = plt.subplots(figsize=(9,3))

ax.scatter(x, np.sin(x))
```

Bar Plot

```python
butter_prices = {'UK': 5.5,
                 'Germany': 4,
                 'Estonia': 2.5,
                 'Sweden': 6.3}

fig, ax = plt.subplots(figsize=(6,4))

ax.bar(butter_prices.keys(), butter_prices.values())

ax.set(title='Butter Prices (EUR)')
```

Horizontal Bar

```python
fig, ax = plt.subplots(figsize=(7,3))

ax.barh(list(butter_prices.keys()), list(butter_prices.values())) # as lists!
```

```python
df = pd.read_csv('car-sales.csv')
df.head()
```

```python
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = df['Price'].str.replace(',','')
df['Price'] = df['Price'].str.replace('.','')
df['Price'] = df['Price'].str[:-2]
df['Price'] = df['Price'].astype(int)
df['Sale Date'] = pd.date_range('1/1/2020', periods=len(df))
```

```python
df['Total Sales'] = df['Price'].cumsum()
df.head()
```

```python
df.plot(x='Sale Date',y='Total Sales')
```

***Plotting Directly From Pandas***

```python
df.plot(x='Make', y='Odometer (KM)', kind='bar') # this is not the average!
```

```python
df = pd.read_csv('heart-disease.csv')
df.head()
```

```python
# Age Distribution
df.age.plot.hist(bins=50)
```

```python
over_50 = df.age[df['age'] > 50]
len(over_50)
```

```python
df[df['age'] > 50].plot(x='age', y='chol', kind='scatter', colormap='winter')
```

```python
# OO method
fig, ax = plt.subplots(figsize=(5,5))

df[df['age'] > 50].plot(kind='scatter',
                        x = 'age',
                        y = 'chol',
                        c='target',
                        ax=ax)

ax.axhline(y=df['chol'].mean(),
           linestyle='--')

# ax.set_xlim([45, 100])
```
